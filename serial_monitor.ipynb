```python
import serial
import time
import csv

# Configura a porta serial
porta = "COM5"   #  troque pela porta correta (ex: COM4 no Windows, /dev/ttyUSB0 no Linux)
baudrate = 115200

# Abre a serial
ser = serial.Serial(porta, baudrate)
time.sleep(2)  # espera estabilizar a comunicao

# Cria o arquivo CSV
with open("data_ecg_2_ms.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["tempo_arduino_ms", "valor_ADC"])  # cabealho

    try:
        while True:
            linha = ser.readline().decode("utf-8").strip()
            if "," in linha:  # confere se veio no formato esperado
                try:
                    tempo_ms, valor = linha.split(",")
                    writer.writerow([tempo_ms, valor])
                    print(tempo_ms, valor)
                except ValueError:
                    # ignora linhas quebradas/ruins
                    pass
    except KeyboardInterrupt:
        print("\nColeta encerrada pelo usurio.")
```
